```r
library(ggpubr)
```

```r
ggscatter(prostate, x = "lweight", y = "gleason", 
          add = "reg.line", conf.int = TRUE, 
          cor.coef = TRUE, cor.method = "pearson",
          xlab = "log(prostate weight)", ylab = "Gleason score")
```

```r
# Shapiro-Wilk normality test for lweight
shapiro.test(prostate$lweight) #  p-value = 6.001e-05
```

```r
# Shapiro-Wilk normality test for wt
shapiro.test(prostate$gleason) # => p-value = 1.049e-12
```

```r
ggqqplot(prostate$lweight, ylab = "Prostate Weight")
```

```r
ggqqplot(prostate$gleason, ylab = "Gleason Score")
```

```r
CM_P3 <- cor.test(prostate$gleason, prostate$lweight, 
                method = "pearson")
CM_P3
```

```r
CM_P3$estimate # --> cor: -0.001283003
```

```r
ggscatter(prostate, x = "gleason", y = "pgg45", 
          add = "reg.line", conf.int = TRUE, 
          cor.coef = TRUE, cor.method = "pearson",
          xlab = "Gleason Score", ylab = "percentage Gleason scores 4 or 5")
```

```r
# Shapiro-Wilk normality test for gleason
shapiro.test(prostate$gleason) #  p-value = 1.049e-12
```

```r
# Shapiro-Wilk normality test for pgg45
shapiro.test(prostate$pgg45) # => p-value = 2.065e-09
```

```r
ggqqplot(prostate$gleason, ylab = "Gleason Score")
```

```r
ggqqplot(prostate$pgg45, ylab = "percentage Gleason scores 4 or 5")
```

```r
CM_P4 <- cor.test(prostate$pgg45, prostate$gleason, 
                  method = "pearson")
CM_P4
```

```r
CM_P4$estimate # --> cor: 0.7519045 
```
